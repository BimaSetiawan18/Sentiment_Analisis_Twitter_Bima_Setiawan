```python
from google_play_scraper import Sort, reviews_all
import csv
import pandas as pd
```

```python
#Mengambil ulasan dari site
scrapreview = reviews_all(
    'com.twitter.android',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=150000
)

scrapreview_limited = scrapreview[:150000]
```

```python
dataView = pd.DataFrame(scrapreview_limited)
dataView.head()
```

```python
with open('hasil_scraping_Twitter.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    writer.writerow(['content','score'])


    for review in scrapreview_limited:
        writer.writerow([
            review['content'],
            review.get('score', '')
        ])

print("Data telah disimpan ke hasil_scraping_Twitter.csv")
```
